```python
# STEP 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# STEP 2: Upload files in Colab (run this cell to upload manually)


# STEP 3: Load the main population dataset
data_file = "API_SP.POP.TOTL_DS2_en_csv_v2_447796.csv"
df = pd.read_csv(data_file, skiprows=4)

# STEP 4: Filter for India
india_df = df[df["Country Name"] == "India"].set_index("Country Name")

# View available years
print("Available Years:", india_df.columns[4:].tolist())

# STEP 5: Get latest population value (2022 for example)
population_2022 = india_df["2022"].values[0]
print("India's Total Population in 2022:", population_2022)

# --- Age Group Data (from UN / your slide example) ---
age_groups = ["0-20 Years", "21-64 Years", "65+ Years"]
population_values = [512, 807, 98]  # in millions (from your provided slide)

# STEP 6: Plot Bar Chart
plt.figure(figsize=(8,6))
plt.bar(age_groups, population_values, color=['gold', 'dodgerblue', 'violet'])
plt.title("India's Population Distribution by Age (2022)")
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
plt.show()
```
